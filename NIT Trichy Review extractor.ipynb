```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
reviewlist = []
```

```python
def get_soup(url):
    data  = requests.get(url).text 
    soup = BeautifulSoup(data,"html.parser")
    return soup
```

```python
def get_reviews(soup):
    reviews = soup.find_all('div',{'class':'cardBlkInn'})
    try:
        for items in reviews:
            review = {
            'Review_title' : items.find('h4',{'class':'blockHeading'}).text.strip(),
            #'Ratings' : float(items.find('img')['alt'][0]),
            'Body' : items.find('div',{'class':'ratingOuter'}).text.strip()
            }
            reviewlist.append(review)
    except:
        pass
```

```python
url = ['https://www.careers360.com/university/national-institute-of-technology-tiruchirappalli/reviews',
      'https://www.careers360.com/university/national-institute-of-technology-tiruchirappalli/reviews?page=2',
      'https://www.careers360.com/university/national-institute-of-technology-tiruchirappalli/reviews?page=3',
      'https://www.careers360.com/university/national-institute-of-technology-tiruchirappalli/reviews?page=4',
      'https://www.careers360.com/university/national-institute-of-technology-tiruchirappalli/reviews?page=5',
      'https://www.careers360.com/university/national-institute-of-technology-tiruchirappalli/reviews?page=6',
       'https://www.careers360.com/university/national-institute-of-technology-tiruchirappalli/reviews?page=7',
      'https://www.careers360.com/university/national-institute-of-technology-tiruchirappalli/reviews?page=8']
```

```python
for i in url:
    soup = get_soup(i)
    get_reviews(soup)
```
